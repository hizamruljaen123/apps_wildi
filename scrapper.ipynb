```python
import requests

def search_products(api_key, query):
    params = {
        'engine': 'google_shopping',  # Menggunakan Google Shopping sebagai mesin pencarian
        'q': query,
        'hl': 'id',  # Bahasa Indonesia
        'gl': 'id',  # Lokasi Indonesia
        'api_key': api_key
    }
    
    response = requests.get('https://serpapi.com/search', params=params)
    data = response.json()

    # Memfilter hasil untuk hanya menampilkan produk dari Tokopedia dan Shopee
    filtered_results = []
    for product in data.get('shopping_results', []):
        source = product.get('source', '').lower()
        if 'tokopedia' in source or 'shopee' in source:
            filtered_results.append(product)
    
    return filtered_results

# Masukkan API Key Anda
api_key = '428ae96fc7601fc90271b22567efd97af8d5ece49b266319797a6fd5551d3a8b'
kata_kunci = 'laptop 256GB SSD 16GB RAM'

# Panggil fungsi pencarian
hasil = search_products(api_key, kata_kunci)

# Menampilkan hasil
for produk in hasil:
    print(f"Nama Produk: {produk['title']}")
    print(f"Harga: {produk['price']}")
    print(f"Sumber: {produk['source']}")
    print(f"Link: {produk['link']}")
    print('-' * 40)
```
